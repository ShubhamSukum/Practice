## PDF to EXCEL

```python
import tabula
import pandas as pd
```

```python
pdfLoc="Placement-Report_2021-22.pdf"
```

```python
tables=tabula.read_pdf(pdfLoc,pages="all",multiple_tables=True)
```

```python
excel_sheet=pd.ExcelWriter("initial2022.xlsx")
```

```python
for i, table in enumerate(tables):
    table.to_excel(excel_sheet,sheet_name=f'Table_{i+1}',index=False)
```

```python
excel_sheet.save()
```
