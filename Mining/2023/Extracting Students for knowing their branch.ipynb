```python
import pandas as pd

# Specify the path to your Excel file (replace 'your_excel_file.xlsx' with your actual file name)
excel_file_path = 'output_file.xlsx'

# Read the first sheet to get column names
first_sheet = pd.read_excel(excel_file_path, sheet_name=0)
columns = first_sheet.columns

# Read all sheets into a dictionary of DataFrames
dfs_dict = pd.read_excel(excel_file_path, sheet_name=None)

# Iterate through the dictionary and append data to the first sheet's columns
merged_df = pd.DataFrame(columns=columns)
for sheet_name, sheet_data in dfs_dict.items():
    if sheet_name != list(dfs_dict.keys())[0]:  # Skip the first sheet
        merged_df = pd.concat([merged_df, sheet_data], ignore_index=True)

# Write the merged DataFrame to a new Excel file
merged_df.to_excel('merged_output.xlsx', index=False)
```
