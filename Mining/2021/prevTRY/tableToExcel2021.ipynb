## PDF to EXCEL

```python
import tabula
import pandas as pd
```

```python
pdfLoc="2020-1-2.pdf"
```

```python
tables=tabula.read_pdf(pdfLoc,pages="all",multiple_tables=True)
```

```python
excel_sheet=pd.ExcelWriter("2021try.xlsx")
```

```python
for i, table in enumerate(tables):
    table.to_excel(excel_sheet,sheet_name=f'Table_{i+1}',index=False)
```

```python
excel_sheet.save()
```
