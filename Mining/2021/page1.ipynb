## 'All (UG)', 'CS & IT (UG)', 'E&TC (UG)'

```python
df['branch'] = df['branch'].str.replace(r'All \(UG and PG\)', 'All (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'CE & IT \(UG( and PG)?\)', 'CE & IT (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'CE & IT \(UG( - [A-Za-z]+)?\)', 'CE & IT (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'All \(UG - [A-Za-z]+\)', 'All (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'All \(UG\) and [A-Za-z]+ \(PG\)', 'All (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'All \(UG\) & [A-Za-z]+ \(PG\)', 'All (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'All \(UG\) and [A-Za-z &]+ \(PG\)', 'All (UG)', regex=True)
df["branch"].unique()
```

```python
df['branch'] = df['branch'].str.replace(r'CE \(UG\)', 'CE', regex=True)
df["branch"].unique()
```
